```python
vaca = pd.read_csv('/Users/kscomputer/Desktop/python_api_challenge/python-api-challenge/Output/weather.csv')
vaca
```

```python
lat_lng = vaca[['Lat','Lng']]

humidity = vaca['Humidity'].astype(float)
```

```python
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(lat_lng, weights = humidity,
                                dissipating=False, max_intensity=100, 
                                 point_radius =4)

fig.add_layer(heat_layer)

fig
```
